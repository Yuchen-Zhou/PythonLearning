# Predicting Price & Profit using Sklearn & StatsModel

```python
#import necessary modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read data
df = pd.read_csv('../dataset/hotels.csv')
df.head()
```

```python
#Check for null values
df.isna().sum()
```

```python
df.describe()
```

## Visual Exploratory data analysis

```python
#Create Price_per_area and ROI
df["Price_per_area"] = df['Price in Millions'] / df['Square Meter']*1000
df['ROI'] = df['Profit']/(df['Price in Millions']*1000)

#Visualize EDA: How each city compare to each other
sns.boxplot(x='City', y='ROI', data=df)
plt.yscale('log')
plt.show()
plt.clf()
sns.boxplot(x="City", y='Price_per_area', data=df)
plt.yscale('log')
plt.show()
```

From the boxplot, it seems that hotels in Cologne tends to have bigger ROI and that although the Price per area in all three cities are varies, prices in Cologne are generally cheaper.

```python
#Look at dependencies of each variables
sns.scatterplot(y='Price in Millions', x='Square Meter', data=df, hue='City')
plt.show()
plt.clf()
g = sns.FacetGrid(df, col='City')
g.map(sns.scatterplot, 'Price in Millions', 'Profit')
plt.show()
```

As we can see the scatterplot, the Price vs Area and Profit in Millions both has linear relationship, although Profit vs. Price in Millions have higher variance. This will enable us to predict the price to buy a hotel based on the area in each city and the profit we could expected to get.

## Predicting Price vs. Area in Each City using statsmodels

```python
#Import ols from statsmodels
from statsmodels.formula.api import ols

df.columns=['Profit', 'Price_in_Millions', 'Square_meter', 'City', 'Price_per_area', 'ROI']
model = ols('Price_in_Millions ~ Square_meter * City + 0', data=df).fit()
print(model.params)
print('Rsquared:', model.rsquared)
```

The model has an rsquared of 0.966 which makes it a very good model

```python
#Create Prediction
City = df['City'].unique()
Square_meter = np.arange(0, 6000, 500)
from itertools import product
p = product(Square_meter, City)
explanatory_data = pd.DataFrame(p, columns=['Square_meter', 'City'])
Prediction_data = explanatory_data.assign(Price_in_Millions=model.predict(explanatory_data))
Prediction_data.head(10)
```

```python
#Visualize the prediction
sns.lmplot(x='Square_meter', y='Price_in_Millions', data=df, hue='City', ci=None)
sns.scatterplot(x='Square_meter', y='Price_in_Millions', data=Prediction_data,
                hue='City', markers='D', legend=False)
plt.show()
```

## Predicting profit based on price, city and square meter area using sklearn

```python
sns.scatterplot(y='Profit', x='Price_in_Millions', data=df, hue='City', ci=None)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

df_modified = pd.get_dummies(df, drop_first=True)
x = df_modified[['Price_in_Millions', 'City_Cologne', 'City_Munich', 'Square_meter']].values
y = df_modified['Profit'].values

#Splitting the variable into the test and train data set
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)

mdl = LinearRegression()
mdl.fit(X_train, y_train)
y_pred = mdl.predict(X_test)
prediction_data = pd.DataFrame(X_test,\
        columns=['Price_in_Millions', 'City_Cologne', 'City_Munich', 'Square_meter'])\
        .assign(y_test=y_test, y_pred=y_pred)
prediction_data.head()
```

## How good is the model?

```python
#Visualize Accuracy
sns.lineplot(x='y_test', y='y_pred', data=prediction_data)
plt.plot(prediction_data['y_test'], prediction_data['y_test'])
plt.show()
```

The linear line represent the perfect model, where the y_pred is always accurately predict the y_test. Based on the graph, our prediction is following the trend, but not quite accurately. Let's quantify this.



To get a more accurate calculation of accuracy, I use cross_val_score to minimize the effect of spliting the data into the train and test data.

```python
from sklearn.model_selection import cross_val_score

cv = cross_val_score(mdl, x, y, cv=5)
print(cv)
print('Average Score:', np.mean(cv))
```

The aveage is 0.46, which is small but understands, considering the features of the data each have their own features.




### From this prediction, we are able to predict how much should someone pay to buy a hotel based on the square meter area and the city with an accuracy of about 96%. We could also predict the profit that we will expected to get from buying the holtel with an accuracy of about 46.5%
